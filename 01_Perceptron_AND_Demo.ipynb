#The code

There are many ways of implementing the algorithm.

```python
import numpy as np

#Set the inputs and outputs,
input = np.array([[-1,-1],[-1,1],[1,-1],[1,1]])
output = np.array([-1,-1,-1,1])

#Set the weights, bias and an array for predicted output 
weights = np.array([0,0])
bias = -1
predicted_output = np.zeros(4)
learning_rate = 0.1 #less than 1

def predict(weights,input,bias):

  summation_fn = np.dot(weights,input)+bias

  if summation_fn>0:
    return 1
  else:
    return -1


print("---Before updates----")
print("weights: ", weights)
print("prediction: ", predicted_output)

#You can have an error function and do a while loop while error!=-0

cycleSize = 10  #later stage, we'll use epoch instead of cycle. This is just so that it's intuitive

for cycle in range(cycleSize):

  print(cycle)
  for row in range(len(input)): 

    predicted_output[row] = predict(weights,input[row],bias)
    
    if predicted_output[row] < output[row]:#if we misclassify positive as negative (mistake on positive)
      
      weights = weights + learning_rate*(output[row]*input[row])
      
    elif predicted_output[row] > output[row]:#if we misclassify negative as positive (mistake on negative)
      
      weights = weights - learning_rate*(output[row]*input[row])


  print("cycle: ",cycle,"weights: ",weights,"prediction: ",predicted_output)  

print("---After many cycle of updates----")
print("weights: ", weights)
print("prediction: ", predicted_output)


```
